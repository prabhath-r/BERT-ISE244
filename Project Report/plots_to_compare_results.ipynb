# **Task-1 Plots: Sentiment Analysis on Twitter US Airlines Sentiment**

Base-BERT model

```python
import matplotlib.pyplot as plt
import numpy as np

batch_sizes = [16, 32, 64]
runtimes = [16, 15, 11]
accuracies = [0.667, 0.65, 0.625]

plt.style.use('seaborn-whitegrid')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

fig.suptitle('BERT Model for Sentiment Analysis', fontsize=14, fontweight='bold')

ax1.bar(np.arange(len(batch_sizes)), runtimes, color='#0072B2', width=0.4)
ax1.set_xlabel('Batch Size')
ax1.set_ylabel('Runtime (minutes)')
ax1.set_title('Runtime vs Batch Size')
ax1.set_xticks(np.arange(len(batch_sizes)))
ax1.set_xticklabels(batch_sizes)
ax1.grid(True)

ax2.bar(np.arange(len(batch_sizes)), accuracies, color='#009E73', width=0.4)
ax2.set_xlabel('Batch Size')
ax2.set_ylabel('Accuracy')
ax2.set_title('Accuracy vs Batch Size')
ax2.set_xticks(np.arange(len(batch_sizes)))
ax2.set_xticklabels(batch_sizes)
ax2.set_ylim(0, 1)
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['DistilBERT', 'BERT','RoBERTa']
runtimes = [48, 90, 65]
accuracies = [0.819, 0.778, 0.857]

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle('Model Comparison', fontsize=16, fontweight='bold')

colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

axs[0].bar(models, runtimes, color=colors, alpha=0.8, align='center')
axs[0].set_xlabel('Models', fontsize=12)
axs[0].set_ylabel('Runtime (minutes)', fontsize=12)
axs[0].set_title('Runtime Comparison', fontsize=14, fontweight='bold')

axs[1].bar(models, accuracies, color=colors, alpha=0.8, align='center')
axs[1].set_xlabel('Models', fontsize=12)
axs[1].set_ylabel('Accuracy', fontsize=12)
axs[1].set_title('Accuracy Comparison', fontsize=14, fontweight='bold')

for i, rt in enumerate(runtimes):
    axs[0].text(i, rt + 5, f'{rt}', ha='center', va='bottom', fontweight='bold', fontsize = 12)

for i, acc in enumerate(accuracies):
    axs[1].text(i, acc + 0.01, f'{acc:.3f}', ha='center', va='bottom', fontweight='bold', fontsize = 12)

axs[0].set_ylim(0, max(runtimes) + 20)
axs[1].set_ylim(0, max(accuracies) + 0.1)
axs[0].grid(axis='y')
axs[1].grid(axis='y')

plt.tight_layout()
plt.show()
```
